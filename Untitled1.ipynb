```python
#When we look at the info again, we see that the 'purpose' is categorical. In order for us to get a good read on it,
#we need to transform it by using dummy variables sou our model will be able to understand it. 
```

```python
pd.get_dummies
```

```python
cat_feats = ['purpose']
```

```python
fdf = pd.get_dummies(df,columns=cat_feats,drop_first=True) #fdf = Final DataFrame
```

```python
fdf.head()
```

```python
#Now that we have reorganized our final dataframe, we can train test split this data 
```

```python
#
```

```python
##Train Test Split##
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = fdf.drop('not.fully.paid',axis=1)
y = fdf['not.fully.paid']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
##Now we can train our decision Tree Model##
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtree = DecisionTreeClassifier()
```

```python
dtree.fit(X_train,y_train)
```

```python
#After our training model data is set up, we can start to make some prediction# 
```

```python
predictions = dtree.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
#To compare the above, we should also compare this with a Random Forest Model#
```

```python
##Training the Random Forest Model## 
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rfc = RandomForestClassifier(n_estimators=600)
```

```python
rfc.fit(X_train,y_train)
```

```python
#After our training model data is set up, we can start to make some prediction# 
```

```python
predictions = rfc.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
print(confusion_matrix(y_test,predictions))
```
